```python
import pandas as pd

# Assuming dataset is your DataFrame and 'element_code' is the column of interest
element_code_column = dataset['Element']

# Initialize variables to store the minimum correlation and corresponding year
min_correlation = 1.0  # Initialize with a high value (correlation ranges from -1 to 1)
year_with_min_correlation = None

# Iterate through the columns representing years
for year_column in dataset.columns:
    if year_column.startswith('y'):  # Assuming year columns start with 'y'
        # Calculate the correlation between 'element_code' and the current year column
        correlation = element_code_column.corr(dataset[year_column])

        # Check if this correlation is smaller than the current minimum
        if correlation < min_correlation:
            min_correlation = correlation
            year_with_min_correlation = year_column

# Now, year_with_min_correlation will contain the year with the least correlation with 'element_code'
print(f"The year with the least correlation to 'element_code' is {year_with_min_correlation}.")
print(f"The minimum correlation value is {min_correlation:.3f}.")
```
